```
The Poisson distribution is a discrete probability distribution that expresses the probability of a given number of events occurring in a fixed interval of time or space, when the events happen independently and the average rate (Î») of occurrence is constant.


```



````python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
````

````python
#Import libraries
from scipy.stats import poisson
````

````python
#Calculation
mean=6
x=range(1,20)
#Formula
poiss_value=poisson.pmf(x,mean)
#Print the value
for i in x:
  print("P(X={})={}".format(i,poiss_value[i-1]))



````

````python
#Plot the bargraph
plt.figure(figsize=(12,8))
plt.bar(x,poiss_value, color='darkcyan')
plt.xlabel("X")
plt.ylabel("P(X)")
plt.title("Poisson Distribution")
plt.grid(True)
plt.tight_layout()
plt.show()
````
