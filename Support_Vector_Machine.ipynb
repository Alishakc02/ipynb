````python
#Import the datasets
from sklearn.datasets import load_digits
````

````python
#Load the datasets
digits=load_digits()
````

````python
#Define the dataframe
df=pd.DataFrame(digits.data, digits.target)
df.head()
#This datasets represent the pixel intensity values
#It is used for the classification of the image
````

````python
X_train, X_test, y_train, y_test=train_test_split(digits.data, digits.target, test_size=0.2)
````

````python
len(X_train)
````

````python
len(X_test)
````

````python
model = SVC(kernel='rbf')# for rbf
model.fit(X_train, y_train)
````

````python
score=model.score(X_test, y_test)
print("Accuracy", score)
````

````python
#Now from linear
model = SVC(kernel='linear')
model.fit(X_train, y_train)
````

````python
#Accuracy
score=model.score(X_test, y_test)
print("Accuracy", score)
````

````python
#Tune your model using regularization and gamma parameters and try to come up with the highest accuracy
model = SVC(C=300, gamma='auto')
model.fit(X_train, y_train)
````

````python
score=model.score(X_test, y_test)
print("Accuracy", score)
````
