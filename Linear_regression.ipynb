```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Data
weight = np.array([3.63, 3.02, 3.82, 3.42, 3.59, 2.87, 3.03, 3.46, 3.36, 3.30])
length = np.array([53.1, 49.7, 48.4, 54.2, 54.9, 43.7, 47.2, 45.2, 54.4, 50.4])

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(weight, length)

# Display regression info
print(f"Regression equation: Length = {intercept:.2f} + {slope:.2f} × Weight")
print(f"R-squared: {r_value**2:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Standard error: {std_err:.4f}")

# Predicted values
predicted_length = intercept + slope * weight

# Plot
plt.figure(figsize=(8, 5))
plt.scatter(weight, length, color='blue', label='Actual Data')
plt.plot(weight, predicted_length, color='red', label='Regression Line')
plt.title("Linear Regression: Weight vs Length")
plt.xlabel("Weight (kg)")
plt.ylabel("Length (cm)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
