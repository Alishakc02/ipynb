```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('player_info.csv')
df
```

```python
df.drop(17, inplace=True)
```

```python
df['player_info']
```

```python
#Player information- Removing \n
df['player_info'].str.split('\n')
```

```python
df['player_info'].str.split('Player Information')
```

```python
df['player_info'].str.split(r'\n').str[2]
```

```python
df['Name']=df['player_info'].str.split(r'\n').str[2].replace('-- : --','NG')
df['Name']
```

```python
df['Age'] = (
    df['player_info']
    .str.split(r'\n').str[4]               # take the 5th element (Age field)
    .replace('-- : --', '0')               # replace missing with 0
    .str.replace('Years', '', regex=False) # remove "Years"
    .astype(int)                           # convert to integer
)
df['Age']
```

```python
df['DOB']=pd.to_datetime(df['player_info'].str.split(r'\n').str[6].replace('-- : --','0'),errors='coerce')
df['DOB']
```

```python
df['Address']=df['player_info'].str.split('\n').str[8]
df['Address']
```

```python
df['Address']=df['player_info'].str.split('\n').str[8].replace('-- : --','NG')
df['Address']
```

```python
df['Nationality']=df['player_info'].str.split('\n').str[10].replace('-- : --','NG').str.strip()
df['Nationality']
```

```python
df
```

```python
df.drop(df.columns[0], axis=1, inplace=True)
df
```

```python
df['profesional_info']
```

```python
df['profesional_info'].str.split('Professional Information')
```

```python
df['profesional_info'][0]
```

```python
df['Role']=df['profesional_info'].str.split(r'\n').str[2]
df['Role']
```

```python
df['Role']=df['profesional_info'].str.split(r'\n').str[2].replace('-- : --','NG')
df['Role']
```

```python
df['Batting Style']=df['profesional_info'].str.split(r'\n').str[4]
df['Batting Style']
```

```python
df['Batting Style']=df['profesional_info'].str.split(r'\n').str[4].replace('-- : --','NG')
df['Batting Style']
```

```python
df['Bowler']=df['profesional_info'].str.split(r'\n').str[6]
df['Bowler']
```

```python
df['Bowler']=df['profesional_info'].str.split(r'\n').str[6].replace('-- : --','NG')
df['Bowler']
```

```python
df['Debut Date']=pd.to_datetime(df['profesional_info'].str.split(r'\n').str[10].replace('-- : --','0'),errors='coerce')
df['Debut Date']
```

```python
df['Team']=df['profesional_info'].str.split(r'\n').str[8]
df['Team']
```

```python
df['Team']=df['profesional_info'].str.split(r'\n').str[8].str.replace('-- : --','NG').str.strip()
df['Team']
```

```python
df
```

```python
df.drop(df.columns[0], axis=1, inplace=True)
df
```
